### Q1. Matrix Multiplication

##### (i) Check if matrix multiplication is valid

```python
def isValid(A,B):
  #your code here
    column_A=len(A[0])
    row_B=len(B)
    if column_A!=row_B:
        return 0
    else: 
        return 1
```

##### (ii) Using loops (without using numpy)

```python
def matrix_multiply(A,B):
  #your code here
    ans=[]
    i=0
    while i<len(A):
        row_ans=[]
        j=0
        while j<len(B[0]):
            product = 0
            v=0
            while v <len(A[i]):
                product += A[i][v] * B[v][j]
                v+=1
            row_ans.append(product)
            j+=1
        ans.append(row_ans)    
        i+=1
    return ans    
```

##### (iii) Using numpy

```python
def matrix_multiply_2(A,B):
  #your code here
    ans=np.dot(A,B)
    return ans
```

### Q2. Z-Score Normalisation

Z-score normalization refers to the process of normalizing every value in a dataset such that the mean of all of the values is 0 and the standard deviation is 1.

We use the following formula to perform a z-score normalization on every value in a dataset:

New value = (x – μ) / σ

where:

x: Original value

μ: Mean of data

σ: Standard deviation of data

##### (i) Without using numpy

```python
def mean(x):
  #your code here
    sum=0
    n = len(x)
    for item in x:
        sum+=item
    return sum/n    
```

```python
def standard_deviation(x):
  #your code here
    sum=0
    n = len(x)
    for item in x:
        sum+=(item-mean(x))**2
    sd=(sum/(n))**0.5
    return sd
```

```python
def zscore_normalisation(x):
  #your code here
    ans=[]
    m=mean(x)
    std=standard_deviation(x)
    for item in x:
        zscr=round((item-m)/std,8)
        ans.append(zscr)
    return ans 
```

##### (ii) Using numpy

Numpy has in_built functions for calculating mean and standard deviation

```python
def zscore_normalisation_2(x):
  #your code here
    dataset=np.array(x)
    std = np.std(dataset)
    mean=np.mean(x)
    z_scr=(dataset-mean)/std
    return z_scr
```

### Q3. Sigmoid fn and its derivative

The sigmoid function is a mathematical function that maps any input value to a value between 0 and 1.

It is defined mathematically as s(x) = 1/(1+e^(-x)).

##### (i) Write a fn to implement sigmoid fn

```python
def sigmoidfn(x):
    sgfn=[]
    #your code here
    for i in range(len(x)):
        sgfn.append(1/(1 + np.exp(-1*x[i])))
        
    return np.array(sgfn)
```

##### (ii) Write a fn to implement derivative of sigmoid fn

```python
def derivative(x):
  #your code here
    der_sgfn=[]
    A=sigmoidfn(x)
    
    for i in range(len(x)):
        der_sgfn.append(A[i]*(1-A[i]))
    return np.array(der_sgfn)
```
